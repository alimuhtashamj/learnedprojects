````python
import pandas as pd
import os 
os.makedirs("company_charts", exist_ok=True)
folder_path = r"C:\Users\Ali Muhtasham Javed\Downloads\NIFTY-50 Stock Market Data (2000 - 2021)"
all_data = []
````

````python
for file in os.listdir(folder_path):
    if file.endswith('csv'):
        df = pd.read_csv(os.path.join(folder_path, file))
        df['Company'] = file.replace('.csv','')
        all_data.append(df)
````

````python
df_all = pd.concat(all_data, ignore_index = True)
````

````python
df_all['Date'] = pd.to_datetime(df_all['Date'])
````

### Why Use `.unique()` When Looping Through Companies?

When working with a large stock dataset like NIFTY-50, each company appears multiple times ‚Äî one row per date.  
If you loop through `df['Company']` directly, you'll get **repeated names**, which means:

- ‚ùå You‚Äôll generate multiple graphs for the **same company**.
- ‚ùå It wastes time and memory.

Instead, use:

```python
companies = df['Company'].unique()


````python

companies = df_selected['Company'].unique()
companies 
````


## üìä Pseudocode: Plotting Stock Prices for Each NIFTY-50 Company

---

### 1Ô∏è‚É£ Import Libraries
- Import `matplotlib.pyplot as plt`
- Import `os`

---

### 2Ô∏è‚É£ Prepare Environment
- Set Matplotlib style: `'seaborn-darkgrid'`
- Get unique list of companies from `df_selected['Company']`
- Create a folder named `company_charts/` if it doesn't already exist

---

### 3Ô∏è‚É£ Loop Through Each Company

**For each `company` in the list of unique companies:**

    ‚Üí Filter rows from `df_selected` where `Company == company`
    ‚Üí Store result in a new DataFrame: `df_company`

    ‚Üí Create a new figure (size: 10x5 inches)
    
    ‚Üí Plot:
        - X-axis: `df_company['Date']`
        - Y-axis: `df_company['Close']`
        - Line color: Blue
        - Label: 'Close Price'
    
    ‚Üí Add chart elements:
        - Title: "{company} - Stock Price Over Time"
        - X-label: "Date"
        - Y-label: "Closing Price"
        - Legend
        - `tight_layout()` to prevent overlap

    ‚Üí Save the plot as a `.png` file in `company_charts/` folder
        - File name format: `{company}_stock_plot.png`

    ‚Üí Close the figure to free up memory

---

### ‚úÖ End of Loop

Repeat until all company plots are saved successfully.



````python
drop_rows = df_selected[df_selected['Company'] == 'stock_metadata'].index
df_selected = df_selected.drop(index=drop_rows)
#index to drop rows. drop() is for columns 
````

````python
# Ensure df_selected is a DataFrame before filtering
if isinstance(df_selected, pd.DataFrame):
	df_selected = df_selected[df_selected['Company'] != 'stock_metadata']
elif isinstance(df_all, pd.DataFrame):
	colums_to_use = ['Date', 'Close', 'Turnover', 'Company']
	df_selected = df_all[colums_to_use].copy()
	df_selected = df_selected[df_selected['Company'] != 'stock_metadata']
else:
	raise ValueError("df_selected is not a DataFrame. Please check the data processing steps.")
````

````python
import matplotlib.pyplot as plt 
for company in companies:
    df_company = df_selected[df_selected['Company'] == company]
    plt.figure(figsize=(10,5))
    plt.plot(df_company['Date'], df_company['Close'], label='Close Price', color='black') 
    plt.title(f"{company} - Stock Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.legend()
    plt.tight_layout()
    save_path = f"company_charts/{company}_stock_plot.png"
    plt.savefig(save_path)
    print(f"Saved: {save_path}")

    # Close the plot to free memory
    plt.close()
````
